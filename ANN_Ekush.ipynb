```python
import os
import shutil

def find_folders_starting_with_number(directory):
    folders = [folder for folder in os.listdir(directory) if os.path.isdir(os.path.join(directory, folder)) and folder[0].isdigit()]
    return folders

directory = '/content/'
new_path = '/content/ekush_small/'
folders_starting_with_number = find_folders_starting_with_number(directory)
if os.path.exists(new_path):
  shutil.rmtree(new_path)
os.mkdir(new_path)
```

```python
def move_folders_into_one(source_dirs, destination_dir):
    try:
        # Create destination directory if it doesn't exist
        os.makedirs(destination_dir, exist_ok=True)

        for source_dir in source_dirs:
            # Get the base name of the source directory
            base_name = os.path.basename(source_dir)
            # Construct the destination path
            destination_path = os.path.join(destination_dir, base_name)
            # Move the source directory to the destination directory
            shutil.move(source_dir, destination_path)
            print("Moved {} to {}.".format(source_dir, destination_path))

        print("All folders moved successfully into {}.".format(destination_dir))
    except Exception as e:
        print("Error occurred while moving folders:", e)

# Example usage
source_directories = folders_starting_with_number
destination_directory = new_path

move_folders_into_one(source_directories, destination_directory)
```

```python
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from tensorflow.keras import layers, models, optimizers, initializers
from tensorflow.keras.layers import Dropout, BatchNormalization
from skimage.feature import local_binary_pattern
```

```python
def load_dataset(folder_path):
    images = []
    labels = []

    class_folders = sorted(os.listdir(folder_path))
    for class_folder in class_folders:
        class_path = os.path.join(folder_path, class_folder)
        if os.path.isdir(class_path):
            label = int(class_folder)  # Assuming class folder names are numeric
            image_files = os.listdir(class_path)
            for image_file in image_files:
                image_path = os.path.join(class_path, image_file)
                if image_file.endswith(('.jpg', '.jpeg', '.png')):  # Adjust extensions as needed
                    image = cv2.imread(image_path)
                    images.append(image)
                    labels.append(label)

    return np.array(images), np.array(labels)

def extract_canny_edge_features(images):
    canny_images = []
    for image in images:
        gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
        edges = cv2.Canny(gray, 100, 200)  # Adjust parameters for better edge detection
        canny_images.append(edges)
    return np.array(canny_images)

def extract_lbp_features(images):
    lbp_images = []
    for image in images:
        gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
        lbp = local_binary_pattern(gray, 8, 1, method='uniform')  # Adjust parameters as needed
        lbp_images.append(lbp)
    return np.array(lbp_images)
```

```python
# Load dataset
folder_path = '/content/ekush_small'
X, y = load_dataset(folder_path)
```

CANNY BASED

```python
X_canny = extract_canny_edge_features(X)

X_flat = X_canny.reshape(X_canny.shape[0], -1)


X_normalized = X_flat / 255.0  # Normalize pixel values to range [0, 1]

X_train, X_test, y_train, y_test = train_test_split(X_normalized, y, test_size=0.2, random_state=42)
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.2, random_state=42)

fig, axes = plt.subplots(nrows=2, ncols=5, figsize=(15, 6))
for i, ax in enumerate(axes.flat):
    ax.imshow(X_canny[i], cmap='gray')
    ax.set_title(f'Class: {y[i]}')
    ax.axis('off')
plt.tight_layout()
plt.show()
```

```python
model = models.Sequential([
    layers.Dense(256, activation='relu', kernel_initializer=initializers.HeNormal(), input_shape=(X_train.shape[1],)),
    layers.Dense(128),
    BatchNormalization(),
    layers.Dropout(0.5),  # Add dropout regularization
    layers.Dense(50, activation='softmax')
])

# Compile the model with reduced learning rate
optimizer = optimizers.Adam(learning_rate=0.001)
model.compile(optimizer=optimizer,
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

# Define a learning rate schedule
lr_schedule = optimizers.schedules.ExponentialDecay(
    initial_learning_rate=0.001,
    decay_steps=10000,
    decay_rate=0.9,
    staircase=True)

# Train the model with learning rate schedule
history = model.fit(X_train, y_train, epochs=10, validation_data=(X_val, y_val), batch_size=32)
```

```python
# Plot training and validation accuracy
plt.plot(history.history['accuracy'], label='Training Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.title('Training and Validation Accuracy')
plt.show()
```

```python
# Evaluate the model
test_loss, test_accuracy = model.evaluate(X_test, y_test)
print("Test Accuracy:", test_accuracy)
```
